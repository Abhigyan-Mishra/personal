# UAV Routing Algorithm using Greedy Method(Coordinates)
## Random Speed and Uncertainity Factor Added to the equation

## Initialization

```python
def distance_finder(one,two) :
    [x1,y1,z1] = one  # first coordinates
    [x2,y2,z2] = two  # second coordinates

    return (((x2-x1)**2)+((y2-y1)**2)+((z2-z1)**2))**(1/2)
```

```python
## Points Extractor From file
import numpy as np
TC=[]
infile = "cleaned_file.txt"
cord=[]
with open(infile) as fin:
    for line in fin:
        str=line.split(',')
        str = [float(i) for i in str] 
        TC.append(str)
        str=[]
TC=np.array(TC)
TC = TC[TC[:,1].argsort()]
print(len(TC))
```

```python
S=[ 4.3720020e+00, 6.1830090e+00,2.2725000e-02]
D= [ 8.9369110e+00, 8.9885930e+00,9.7317000e-02]
WP=[[7.849713,12.810686,0.244230],[5.946933,7.345695,0.074365],[5.472060,5.931275,-0.016850],[4.855054,5.494247,-0.014698]]
ST=[] #Source to Task Distance
TD=[] #Task to Destination Distance
TT=[] #Inter Task Distance
TTi=[]
dev=100
TotalD=0
TotalD+=distance_finder(S,WP[0])
for i in range(0,len(WP)-1):
    TotalD+=distance_finder(WP[i],WP[i+1])
#TotalD+=distance_finder(WP[len(W)-1],D)
print("Total Distance on optimal path:",TotalD)
print("Total Deviation allowed from optimal path:",dev)
for i in range(0,len(TC)):
    dist=distance_finder(S,TC[i])
    ST.append(dist)
print("Source to Task distance:",(ST))
for i in range(0,len(TC)):
    dist=distance_finder(TC[i],D)
    TD.append(dist)
print("Task to Destination distance:",TD)
for i in range(0,len(TC)):
    for j in range(0,len(TC)):
        dist=distance_finder(TC[i],TC[j])
        TTi.append(dist)
    TT.append(TTi)
    TTi=[]
print("Distance Between Tasks:",TT)
```

```python
import random
t=[]   #Time it takes to perform each task
rew=[] #Reward for each task
#Random Time
for i in range(len(TC)):
    rew.append(random.randint(5,10))
    t.append(random.randint(5,15))
print("Time:",t)
print("Reward:",rew)
#Function to find random speed
def randSpeed():
    spd=[]
    for i in range(len(TC)):
      spd.append(random.randint(10,20))
    return spd
#Fucntion to get the uncertainity values
def randUnCT():
  unCT=[]
  for i in range(len(TC)):
    unCT.append(round(random.random(),2))
  return unCT
```

```python
Tmax=60              #Total Time Limit
Dmax=TotalD+dev      #Maximum Distance => Optimal_Distance+Deviation_Allowed
totR=0              #Total Reward
TimeST=[] #Time to reach from Source to Task
TimeTT=[] #Time to reach from Task to Task
TimeTD=[] #Time to reach from Task to Destination
#random Time calculations after each step
def randTime():
  TimeST=[] 
  TimeTT=[]
  TimeTD=[]
  spd=randSpeed()
  for i in range(0,len(TC)):  #Calculating TimeST
    temp=ST[i]/spd[i]
    TimeST.append(temp)
  for i in range(0,len(t)): #Calculating TimeTD
    temp=TD[i]/spd[i]
    TimeTD.append(temp)
  for i in range(0,len(TT)):
     temp=[]
     for j in range(0,len(TT[i])):
        temp1=TT[i][j]/spd[i]
        temp.append(temp1)
     TimeTT.append(temp)
  return TimeST,TimeTT,TimeTD
ta=[]               #Choosen Tasks
print(Dmax)
```

```python
#First Task Choice
max=0
eff=0
s=0
for i in range(0,len(ST)):
    if((ST[i]+TD[i])<=Dmax and t[i]<=Tmax):
        unCT=randUnCT()
        TimeST,TimeTT,TimeTD=randTime()
        eff1=rew[i]/(ST[i]+t[i]+TimeST[i])
        eff=unCT[i]*eff1
    print("Task",i," Efficiency:",eff)
    if(eff>max):
        max=eff
        s=i
    eff=0  
print("Task Chossen: ",s)
ta.append(s)
totR+=rew[s]
Tmax=Tmax-(t[s]+TimeST[s])            
Dmax=Dmax-ST[s]
```

```python
#Intermediate Task Selection
max=0
n=s
while(s!=-1):
    TimeST,TimeTT,TimeTD=randTime()
    row=s
    rowTD=TT[row]
    rowTimeTD=TimeTT[row]
    for col in range(s+1,len(ST)):
       if((t[col]+rowTimeTD[col])<=Tmax and (rowTD[col]+TD[col])<=Dmax):
        unCT=randUnCT()
        eff1=rew[col]/(t[col]+rowTD[col]+rowTimeTD[col])
        eff=unCT[i]*eff1
        print("Task :",col," ",eff)
        if(eff>max):
          max=eff
          s=col
    max=0   
    if(s>row and s<len(ST)):     
      print(max)
      print("Task:",s)
      print("Remaining Time:",(Tmax-(t[s]+TimeTT[n][s])))
      Dtot=TT[n][s]
      print("Remaining Deviation:",(Dmax-Dtot))
      totR=totR+rew[s]
      print("Reward:",totR)
      Tmax=Tmax-(t[s]+TimeTT[n][s])
      Dmax=Dmax-Dtot
      ta.append(s)
      n=s
    else:
     s=-1
Dmax=Dmax-TD[s-1]
Tmax=Tmax-TimeTD[s-1]
```
